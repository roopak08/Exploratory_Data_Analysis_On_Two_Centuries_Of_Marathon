```python
#import libraries
```

```python
import pandas as pd
```

```python
import seaborn as sb
```

```python
df = pd.read_csv("TWO_CENTURIES_OF_UM_RACES.csv")
```

```python
df.head(10)
```

```python
df.shape
```

```python
df.dtypes
```

```python
#cleaning data 
```

```python
#chosing country specific - USA (50k or 50 miles in 2020)
```

```python
#step 1 - show 50k or 50 miles
#50k
#50mi
#50miles
```

```python
df[df['Event distance/length'] == '50miles']
```

```python
#combine 50k, 50mi, 50miles with isin
df[(df['Event distance/length'].isin(['50mi', '50km', '50miles'])) & (df['Year of event'] == 2020)]
```

```python
df[df['Event name'] == 'Everglades 50 Mile Ultra Run (USA)']['Event name'].str.split('(').str.get(1).str.split(')').str.get(0)
```

```python
df[df['Event name'].str.split('(').str.get(1).str.split(')').str.get(0) == 'USA']
```

```python
#combine all the filters together
```

```python
df2 = df[(df['Event distance/length'].isin(['50mi', '50km', '50miles'])) & (df['Year of event'] == 2020)  & (df['Event name'].str.split('(').str.get(1).str.split(')').str.get(0) == 'USA')]
```

```python
df2.head(10)
```

```python
df2.shape
```

```python
#remove USA from Event name
df2['Event name'].str.split('(').str.get(0)
```

```python
df2['Event name'] = df2['Event name'].str.split('(').str.get(0)
```

```python
df2.head(10)
```

```python
#cleaning athlete
```

```python
df2['athlete age'] = df['Year of event'] - df['Athlete year of birth']
```

```python
df2.head(10)
```

```python
#remove h from atlete performance
```

```python
df2['Athlete performance'] = df2['Athlete performance'].str.split(' ').str.get(0)
```

```python
#drop athlete clubs from the data
```

```python
df2 = df2.drop(['Athlete club', 'Athlete country', 'Athlete year of birth', 'Athlete age category'], axis=1)
```

```python
df2.head(5)
```

```python
#clean null values
```

```python
df2.isna().sum()
```

```python
df2[df2['athlete age'].isna() == 1]
```

```python
df2 = df2.dropna()
```

```python
df2.isna().sum()
```

```python
#check for duplicate values
```

```python
df2[df2.duplicated() == True]
```

```python
#reseting index
```

```python
df2.reset_index(drop = True)
```

```python
# fix types
```

```python
df2.dtypes
```

```python
df2['athlete age'] = df2['athlete age'].astype(int)
```

```python
df2['Athlete average speed'] = df2['Athlete average speed'].astype(float)
```

```python
df2.dtypes
```

```python
df2.head(5)
```

```python
#rename columns for easer usage
```

```python
df2 = df2.rename(columns ={'Year of event': 'event_year', 'Event dates': 'event_date', 'Event name': 'event_name', 'Event distance/length': 'event_distance', 'Event number of finishers': 'event_finishers', 'Athlete performance': 'athlete_performance', 'Athlete gender': 'athlete_gender', 'Athlete average speed': 'athlete_average_speed', 'Athlete ID': 'athlete_id', 'athlete age': 'athlete_age'})
```

```python
df2.head(5)
```

```python
# reorder columns
```

```python
df3 = df2[['event_date', 'event_name', 'event_distance', 'event_finishers', 'athlete_id', 'athlete_gender', 'athlete_age', 'athlete_performance', 'athlete_average_speed']]
```

```python
df3.head(5)
```

```python
# charts and graphs
```

```python
sb.histplot(df3['event_distance'])
```

```python
sb.histplot(df3, x = 'event_distance', hue = 'athlete_gender')
```

```python
sb.displot(df3[df3['event_distance'] == '50mi']['athlete_average_speed'])
```

```python
sb.violinplot(data = df3, x='event_distance', y = 'athlete_average_speed', hue = 'athlete_gender', split = True, inner = 'quart', linewidth=1)
```

```python
sb.lmplot(data = df3, x = 'athlete_age', y = 'athlete_average_speed', hue = 'athlete_gender')
```

```python
# difference in speed for the 50k, 50mi male to female
```

```python
df3.groupby(['event_distance', 'athlete_gender'])['athlete_average_speed'].mean()
```

```python
# what age group are best in 50mi race (20+ race min)
```

```python
df3.query('event_distance == "50mi"').groupby('athlete_age')['athlete_average_speed'].agg(['mean', 'count']).sort_values('mean', ascending = False).query('count>19').head(10)
```

```python
# what age group are worst in 50mi race (10+ race min)
```

```python
df3.query('event_distance == "50mi"').groupby('athlete_age')['athlete_average_speed'].agg(['mean', 'count']).sort_values('mean', ascending = True).query('count>9').head(10)
```

```python
# Seasons for the data -> Slower in summer than winter??

# sring 3-5
# summer 6-8
# fall 9-11
# winter 12- 2
```

```python
df3['event_month'] = df3['event_date'].str.split('.').str.get(1).astype(int)
```

```python
df3.head(5)
```

```python
df3['event_season'] = df3['event_month'].apply(lambda x: 'Winter' if x > 11 else 'Fall' if x > 8 else 'Summer' if x > 5 else 'Spring' if x> 2 else 'Winter')
```

```python
df3.head(25)
```

```python
df3.groupby('event_season')['athlete_average_speed'].agg(['mean', 'count']).sort_values('mean', ascending = False)
```

```python
# 50 miler only
```

```python
df3.query('event_distance == "50mi"').groupby('event_season')['athlete_average_speed'].agg(['mean', 'count']).sort_values('mean', ascending = False)
```
